```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t0 = 0        # start time
tf = 2*np.pi  # end time
dt = 0.1      # step-size
```

```python
t = np.arange(t0, tf, dt)  # 기록 간격
```

```python
print(t)
```

```python
# Plot Results
plt.plot(t, np.cos(t), 's' , markersize=8, markerfacecolor = 'none', markeredgecolor='red', markeredgewidth=2.5,  label="Analytic y=cos(t)")
#  plt.plot(x axis, cosine, label)
plt.plot(t, np.sin(t), "b",  label="Analytic y=sin(t)")

plt.xlabel("Time (t)")
plt.ylabel("fuction value (y)")
plt.title("Graphs")

plt.xticks(rotation=45)
# x axis ticks sengetkan sikit for neater visual

plt.grid()
plt.legend()

plt.show()
```
